# 4.Show the number of players in each country

```python
pd.set_option('display.max_rows', None)
players_df = fifa_df.groupby("nationality").size().to_frame(name='players')
players_df.index.names = ['country']
players_df
```

# 5.Show the top 10 countries with the highest amount of players

```python
most_players_df = players_df.nlargest(10, 'players').reset_index()
most_players_df.index += 1
most_players_df
```

# 6.Create a bar plot of the top 5 countries with the highest number of players, and fill the bars in green

```python
ax = sns.barplot(x = 'country', y = "players", data = most_players_df.head(), color = 'green')
ax.figure.set_size_inches(20, 8)
ax.set_title('Top 5 Countries with Most Players', fontsize = 18)
ax.set_ylabel('Amount of Players', fontsize = 12)
ax.set_xlabel('Country', fontsize = 12)
plt.show()
```

# 7.Display the short name and wages for the top 5 players

```python
fifa_df.head()
```

```python
fifa_df['short_name'] = fifa_df['short_name'].replace('Cristiano Ronaldo','C. Ronaldo')
fifa_df.head()
```

```python
top5_players = fifa_df.head()[['short_name', 'wage_eur']].reset_index(drop=True)
top5_players.index += 1
top5_players
```

# 8.Display the short name and wages for the top 5 players with the highest salaries

```python
top5_players_wage = fifa_df[['short_name', 'wage_eur']].nlargest(5, 'wage_eur').reset_index(drop=True)
top5_players_wage.index += 1
top5_players_wage
```

# 9.Create a bar plot of the previous bullet point

```python
ax = sns.barplot(x = 'short_name', y = "wage_eur", data = top5_players_wage)
ax.figure.set_size_inches(20, 8)
ax.set_title('Top 5 Players with Highest Salaries', fontsize = 18)
ax.set_ylabel('Wage (Euros)', fontsize = 12)
ax.set_xlabel('Player', fontsize = 12)
plt.show()
```
